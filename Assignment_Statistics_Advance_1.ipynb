1. What is a random variable in probability theory?
   - A random variable is a function that maps outcomes of a sample space to real numbers.

   Example:

   * Suppose we toss a fair coin:
     
     * Sample space S = {Heads, Tails}

     * Define a random variable X such that:
    
        * X(Heads) = 1

        * X(Tails) = 0

   Than x is a random variable than assign number each outcome.

   * Types of Random Variables:

      1. Discrete Random Variable

         Takes on countable values (e.g., 0, 1, 2, ...).

      2. Continuous Random Variable

         Takes on uncountable/infinite values within a range.

2. What are the types of random variables?
   - There are two main types of random variables in probability theory:

     1. Discrete Random Variable
        
        A discrete random variable takes on a finite or countably infinite set of possible values.

        * characteristics:
          
          . Values are distinct and separate (like 0, 1, 2, ...)

          . Often arise from counting (e.g., number of headss , number of students)

          . Examples:
            
            . Tossing a coin 3 times -> x = numbers of heights -> possible values: 0, 1,2,3

            . Rolling a die -> X = outcome -> possible values: 1,2,3,4,5,6

          . Uses:
            
            . Probability Mass Function (PMF)
              describe the probability of each value.

      2. Continues random variable
         
         A discrete random variable takes on a flnite or countable infinite set of possible values.

         * Characteristics:
           
           . Can take any value in a real interval

           . Often arise from measuring (e.g., height, time, weight)

         * Examples:
             
             . Time taken to run a race -> X could be 9.85, 10.12, etc.

             . Temperature on a day -> X = 25.6°C

         * Uses:
            
            . Probability Density Function (PDF) is used, since the probability of any exact value is 0.

            . Probabilities are calculated over intervals (e.g., p(10<X< 15))

21. What are the assumptions for applying the Central Limit Theorem?
    - The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution, regardless of the population’s shape, provided certain conditions are met.

    1. Random Sampling - Samples must be drawn randomly and independently from the population.

    2. Sample Size - Larger sample sizes (typically n ≥ 30) make the CLT more reliable.

    3. Finite Mean and Variance - The population must have a finite mean μ and finite variance σ2.

    4. Independent Observations - Each observation in the sample should be independent of the others.

    5. Identically Distributed (ideal) - 	Ideally, each data point comes from the same distribution (though some variation is tolerated in practice).
